```python
import pandas as pd
import ast
```

```python
data = pd.read_csv('tmdb_5000_movies.csv')
data.head()
```

```python
data.columns
```

```python
# Dropping the unnecessary columns
list = ['budget', 'homepage', 'original_language', 'original_title', 'popularity', 'production_companies',
       'production_countries', 'release_date', 'revenue', 'runtime',
       'spoken_languages', 'status', 'vote_average',
       'vote_count', 'tagline']

data = data.drop(list, axis = 1)
```

```python
data.head()
```

```python
data.shape
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
credits.head()
```

```python
credits = credits.drop(['movie_id'], axis = 1)
credits.head()
```

```python
# Merging both the dataset on the basis of the 'title'
data = data.merge(credits, on='title')
data.head()
```

Data Preprocessing

```python
# Dropping the rows which have null value in 'overview'
data.dropna(subset=['overview'], inplace=True)
```

```python
# We only need the genres and not the other items in the list
def getval(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
data['genres'] = data['genres'].apply(getval)

data.head()
```

```python
data['keywords'] = data['keywords'].apply(getval)
data.head()
```

```python
# We only want the top three characters in each movie, hence we apply the below function
def getChar(obj):
    count = 0
    list = []
    for i in ast.literal_eval(obj):
        if count < 3:
            list.append(i['name'])
            count+=1
    return list
data['cast'] = data['cast'].apply(getChar)

data.head()
```

```python
def getDirector(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
    return list
data['crew'] = data['crew'].apply(getDirector)

data.head()
```

```python
main = data.copy()
```

```python
data['cast'] = data['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
data['genres'] = data['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
data['keywords'] = data['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
data['crew']= data['crew'].apply(lambda x: [i.replace(" ", "") for i in x])

data.head()
```

```python
data['overview'] = data['overview'].apply(lambda x: x.split())
data.head()
```

```python
# Combining keywords, overview, title, cast, crew, genres into 1 column
data['tag'] = data['overview'] + data['cast'] + data['cast'] + data['crew'] + data['genres']
data = data.drop(['keywords', 'overview', 'cast', 'crew', 'genres'], axis = 1)
data.head()
```

```python
data['tag'] = data['tag'].apply(lambda x: " ".join(x))
data.head()
```

```python
data['tag'] = data['tag'].apply(lambda x: x.lower())
data.head()
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)

data['tag'] = data['tag'].apply(stem)
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(data['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    index = data[data['title'] == movie].index[0]
    distances = sorted(enumerate(similarity[index]), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(data.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
